```python
import os
import pandas as pd
import xml.etree.ElementTree as ET

def rename_files_and_update_xml(image_dir, xml_dir, output_index_file):
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('png', 'jpg', 'jpeg', 'tif', 'bmp'))]
    xml_files = [f for f in os.listdir(xml_dir) if f.lower().endswith('xml')]

    old_new_names = []

    for idx, (image_file, xml_file) in enumerate(zip(image_files, xml_files)):
        old_image_path = os.path.join(image_dir, image_file)
        old_xml_path = os.path.join(xml_dir, xml_file)

        new_base_name = f'image_{idx:04d}'
        new_image_name = f'{new_base_name}{os.path.splitext(image_file)[1]}'
        new_xml_name = f'{new_base_name}.xml'

        new_image_path = os.path.join(image_dir, new_image_name)
        new_xml_path = os.path.join(xml_dir, new_xml_name)

        os.rename(old_image_path, new_image_path)
        os.rename(old_xml_path, new_xml_path)

        # Update the XML file with the new image filename
        tree = ET.parse(new_xml_path)
        root = tree.getroot()
        ns = {'ns': 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2019-07-15'}
        page_element = root.find('.//ns:Page', ns)
        if page_element is not None:
            page_element.attrib['imageFilename'] = new_image_name
            tree.write(new_xml_path)

        old_new_names.append([image_file, new_image_name])

    # Save the old and new names to a DataFrame and then to a CSV file
    df = pd.DataFrame(old_new_names, columns=['Old Name', 'New Name'])
    df.to_csv(output_index_file, index=False)
    print(f'Index file saved to {output_index_file}')

# Example usage:
image_dir = 'training_data/page'
xml_dir = 'training_data/page/'
output_index_file = 'training_data/filename_index.csv'

rename_files_and_update_xml(image_dir, xml_dir, output_index_file)
```
